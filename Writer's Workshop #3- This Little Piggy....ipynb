# Writer's Workshop #3- This Little Piggy...

The written Japanese language has three different 'scripts'.

Usually kanji doens't have any pronounciation- you are expected to know what it means and how it's prounouced.

However, for children or new language learners, prounciation aid can help a lot.

There is a HTML markup to add these pronounciation aids. It is called 'ruby' (confusingly enough, no, it's not the programming language Ruby).

Let's try 'ruby' out.

First, we need to import HTML and display it, since JupyterLab uses Python.

```python
from IPython.core.display import display, HTML
```

Next, let's display some kanji

```python
display(HTML('<h1>焼きそば</h1>'))
```

As you can see, the first character is kanji, and the other three are hiragana. If I can only read hiragana, I have no idea what the full word means. :(

So for me, I can read 'something-ki-so-ba'

If someone were to add prounciation aid by 'ruby' HTML markup, I can read the full word.

```python
from IPython.core.display import display, HTML

display(HTML('<h1><ruby>焼きそば <rp>（</rp><rt>やきそば </rt><rp>）</rp></ruby></h1>'))
```

Oh! The word is... ya-ki-so-ba!! Delicious!


Now's let's play around with 'ruby' for other uses! Let's try using English text!

```python
from IPython.core.display import display, HTML

display(HTML('<h1><ruby>焼きそば <rp>（</rp><rt>abcdfg </rt><rp>）</rp></ruby></h1>'))

display(HTML('<h1><ruby>This little piggy went to the market! <rp>（</rp><rt>やきそば </rt><rp>）</rp></ruby></h1>'))
```

Now, what if we changed it to emoji, the next frontiner language?

```python
display(HTML('<h1><ruby>This little piggy went to the market! <rp>（</rp><rt>や🐖そば </rt><rp>）</rp></ruby></h1>'))
```

Yay! Let's continue and finish the nursery rhyme!
